# Preprocerssing on texual data

```python
import nltk
```

```python
import re
from nltk.corpus import stopwords
sw = nltk.corpus.stopwords.words('english')
```

```python
sw.remove("not")
```

```python
from nltk.stem.wordnet import WordNetLemmatizer          
wn = nltk.WordNetLemmatizer()
```

```python
corpus=[]
```

```python
for i in range(1000):
    review=data["Review"][i]
    review=re.sub("[^a-zA-Z]"," ",review)
    review=review.lower()
    review=review.split()
    d=[]
    for word in review:
        if (word not in sw):
            d.append(wn.lemmatize(word))
    d=" ".join(d)
    corpus.append(d)
```

```python
corpus
```

Data is cleaned now

# Vectorization(TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vec=TfidfVectorizer()
```

```python
x_tf=tfidf_vec.fit_transform(corpus).toarray()
x_tf.shape
```

```python
x_tf
```

# Spliting of data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_tf,data["Liked"], test_size = 0.30)
```

# Building NN

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Building Model
model=Sequential()
```

# Input layer

```python
x_train.shape[1]
```

```python
y_train.shape
```

```python
model.add(Dense(input_dim=x_train.shape[1],kernel_initializer="random_uniform",activation="relu",units=884))
```

# Hidden layer

```python
model.add(Dense(kernel_initializer="random_uniform",activation="relu",units=442))
```

# Output layer

```python
model.add(Dense(kernel_initializer="random_uniform",activation="sigmoid",units=1))
```

# compile data

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
train_history = model.fit(x_train,y_train,epochs=10,batch_size=32)
```

# Prediction

```python
y_pred=model.predict_classes(x_test)
y_pred
```

# Accuaracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Save model

```python
model.save("sentiment_analysis.h5")
```
